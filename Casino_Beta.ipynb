```python
import random

class Gambling:
    def __init__(self , money):
        self.money = money
        self.luck = 0
    
    def get_money(self):
        return self.money
    
    def set_money(self , amt):
        self.money = amt
    
    def add_money(self , amt):
        self.money += amt
    
    def get_luck(self):
        return "Your luck is {}, donate to charity for better luck".format(self.luck) 
    
    def draw_suit_num(self , n):
        suits = ["Diamonds" , "Clubs" , "Hearts" , "Spades" ]
        num = ["Ace" , "2" , "3" , "4" , "5" , "6" , '7' , '8' , '9' , '10' , "Jack" , "Queen" , "King"]
        string = []
        for item in suits:
            for item1 in num:
                string.append("{} of {}".format(item1 , item))
        lst = random.sample(range(52) , 52)
        deck = []
        for item in lst:
            deck.append(string[item])
        ans = []
        for i in range(n):
            ans.append(deck[i])
        return ans
    
    def draw_num(self , n):
        num = ["Ace" , "2" , "3" , "4" , "5" , "6" , '7' , '8' , '9' , '10' , "Jack" , "Queen" , "King"]
        lst = random.sample(range(13) , n)
        deck = []
        for item in lst:
            deck.append(num[item])
        return deck


    def blackjack(self , bet_amt):
        def card_value(card):
            if card in ['Jack', 'Queen', 'King']:
                return 10
            elif card == 'Ace':
                return 1 
            else:
                return int(card)
        
        def highest_total(hand):
            ans = 0
            for card in hand:
                if card == "Ace":
                    ans += 11
                else:
                    ans += card_value(card)
            return ans
        
        def lowest_total(hand):
            ans = 0
            for card in hand:
                if card == "Ace":
                    ans += 1
                else:
                    ans += card_value(card)
            return ans
        def draw_numbers(n):
            num = ["Ace" , "2" , "3" , "4" , "5" , "6" , '7' , '8' , '9' , '10' , "Jack" , "Queen" , "King"]
            lst = random.sample(range(13) , n)
            deck = []
            for item in lst:
                deck.append(num[item])
            return deck

        #Original 2 cards
        Player_Hand = draw_numbers(2)
        dealer_hand = draw_numbers(2)
              
            
        #Getting passport
        while highest_total(dealer_hand) < 16:
            dealer_hand.append(draw_numbers(1)[0])
        
        while highest_total(Player_Hand) < 16:
            Player_Hand.append(draw_numbers(1)[0])
        
        print("Dealer: {}".format(dealer_hand))
        print("Player: {}".format(Player_Hand))
        dealer_points = highest_total(dealer_hand)
        
        #Check for instant wins by dealer or player
        if Player_Hand[0] == "Ace" and Player_Hand[1] == "Ace":
            print("Ban-ban, you win {} dollars".format(str(bet_amt * 3)))
            self.money += bet_amt * 3
            return
        elif (Player_Hand[0] == "Ace" and Player_Hand[1] in ["10" , "Jack" , "Queen" , "King"]) or (Player_Hand[1] == "Ace" and Player_Hand[0] in ["10" , "Jack" , "Queen" , "King"]):
            print("Ban-luc, you win {} dollars".format(str(bet_amt * 2)))
            self.money += bet_amt * 2
            return
        elif (dealer_hand[0] == "Ace") and dealer_hand[1] == "Ace":
            print("Dealer ban-ban, lose {} dollars".format(str(bet_amt * 3)))
            self.money -= bet_amt * 3
            return
        elif (dealer_hand[0] == "Ace" and dealer_hand[1] in ["10" , "Jack" , "Queen" , "King"]) or (dealer_hand[1] == "Ace" and dealer_hand[0] in ["10" , "Jack" , "Queen" , "King"]):
            print("Dealer Ban-luc, lose {} dollars".format(str(bet_amt * 2)))
            self.money -= bet_amt * 2
            return
        
        #Other cases where there is no instant win
        
        
        #Five card hand
        if len(Player_Hand) == 5:
            if lowest_total(Player_Hand) <= 21:
                print("Five Dragons! You win {} dollars".format(str(bet_amt*2)))
                self.money += bet_amt * 2
                return
            else:
                print("Five Dragon Bust , you lose {} dollars".format(str(bet_amt*2)))
                self.money -= bet_amt * 2
                return
        #For hands
        else:
            player_points = highest_total(Player_Hand)
            if len(Player_Hand) == 3 and Player_Hand[0] == "7" and Player_Hand[1] == "7" and Player_Hand[2] == "7":
                print("Triple 7, win {} dollars".format(bet_amt * 7))
                self.money += bet_amt * 7
                return
            elif len(dealer_hand) == 3 and dealer_hand[0] == "7" and dealer_hand[1] == "7" and dealer_hand[2] == "7":
                print("Dealer triple 7, lose {} dollars".format(bet_amt * 7))
                self.money -= bet_amt * 7
                return
            
            elif player_points > 21 and dealer_points > 21:
                print("Both bust, lose {} dollars".format(bet_amt))
                self.money -= bet_amt
                return
            elif player_points > 21 and dealer_points < 21:
                print("Bust, lose {} dollars".format(bet_amt))
                self.money -= bet_amt
                return
            elif dealer_points > 21 and player_points < 21:
                print("Dealer bust, win {} dollars".format(bet_amt))
                self.money += bet_amt
                return
            elif dealer_points > player_points:
                print("Dealer wins, lose {} dollars".format(bet_amt))
                self.money -= bet_amt
                return
            elif dealer_points < player_points:
                print("Dealer loses, win {} dollars".format(bet_amt))
                self.money += bet_amt
                return
            else:
                print("Draw, no change")
                return
                    
            
                
        
            
        

```
